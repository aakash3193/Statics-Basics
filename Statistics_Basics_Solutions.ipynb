## Question 1
**What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

**Answer:**
- **Descriptive statistics** It consist of organizing and summarizing complete population/data (e.g., Strike rate of Virat Kohli in ODI; Avg age of students in a class).
- **Inferential statistics** draw population conclusions from samples with uncertainty (e.g., Avg weight of population in India, Number of trees in India).

## Question 2
**What is sampling in statistics? Explain the differences between random and stratified sampling.**

**Answer:**
- **Sampling** selects a subset of a population for measurement.
- **Simple random sampling (SRS):** every unit has equal chance; works well when population is fairly homogeneous.
- **Stratified sampling:** split into strata (homogeneous subgroups) and randomly sample within each; improves precision and ensures subgroup coverage (e.g., by age or region).

## Question 3
**Define mean, median, and mode. Explain why these measures of central tendency are important.**

**Answer:**
- **Mean:** arithmetic average; sensitive to outliers.
- **Median:** physical mid point of the data; robust to outliers and skew.
- **Mode:** most frequent value(s); 

In all of these cases we are trying to represent the data with a central value. These summarize where data “typically” lie and support comparison and modeling.

## Question 4
**Explain skewness and kurtosis. What does a positive skew imply about the data?**

**Answer:**
- **Skewness:** asymmetry. Positive (right) skew ⇒ long right tail; mean > median; a few high values.
- **Kurtosis:** tail heaviness/peakedness (excess kurtosis > 0 ⇒ heavier tails).
**Positive skew** means most values are lower with a few large ones stretching the right tail.

## Question 6
**Covariance and correlation for:**
`list_x = [10, 20, 30, 40, 50]`

`list_y = [15, 25, 35, 45, 60]`

```python
import numpy as np
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]
x = np.array(list_x, dtype=float)
y = np.array(list_y, dtype=float)
print('Covariance (sample):', np.cov(x, y, ddof=1)[0, 1])
print('Correlation (Pearson r):', np.corrcoef(x, y)[0, 1])
```

## Question 7
**Boxplot and outliers for:**
`data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]`

Steps: draw boxplot → compute IQR → outliers via 1.5×IQR rule.

```python
import numpy as np
import matplotlib.pyplot as plt
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]
arr = np.array(data, dtype=float)
Q1 = np.percentile(arr, 25)
Q3 = np.percentile(arr, 75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
outliers = arr[(arr < lower_fence) | (arr > upper_fence)]
print('Q1:', Q1, 'Q3:', Q3, 'IQR:', IQR)
print('Lower fence:', lower_fence, 'Upper fence:', upper_fence)
print('Outliers:', list(outliers))
plt.figure()
plt.boxplot(arr)
plt.title('Boxplot with IQR-based Outliers')
plt.ylabel('Value')
plt.show()
```

## Question 8
**E-commerce: Is daily sales related to advertising spend?**

**Approach:**
- **Covariance** → direction (positive/negative).
- **Correlation (Pearson r)** → direction + standardized strength (−1 to 1).
Also plot a scatter; consider simple linear regression if r is strong.

```python
import numpy as np
import matplotlib.pyplot as plt
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]
a = np.array(advertising_spend, dtype=float)
s = np.array(daily_sales, dtype=float)
print('Covariance (sample):', np.cov(a, s, ddof=1)[0, 1])
print('Correlation (Pearson r):', np.corrcoef(a, s)[0, 1])
plt.figure()
plt.scatter(a, s)
plt.xlabel('Advertising Spend')
plt.ylabel('Daily Sales')
plt.title('Advertising vs Daily Sales')
plt.show()
```

## Question 9
**Customer satisfaction (1–10): summary stats + histogram**

Use count, mean, median, mode(s), std, min/max; visualize with histogram (and optionally boxplot).

```python
import numpy as np
from statistics import mean, median, multimode
import matplotlib.pyplot as plt
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]
arr = np.array(survey_scores, dtype=float)
print('count:', arr.size)
print('mean:', mean(survey_scores))
print('median:', median(survey_scores))
print('mode(s):', multimode(survey_scores))
print('std (sample):', arr.std(ddof=1))
print('min:', arr.min(), 'max:', arr.max())
plt.figure()
plt.hist(arr, bins=range(int(arr.min()), int(arr.max())+2))
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Survey Score Distribution')
plt.show()
```
